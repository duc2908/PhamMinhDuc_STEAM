```python
train_txt_path = "/workspace/nabang1010/STEAM/PhamMinhDuc_STEAM/src/yolov7/trainset.txt"
train_txt = open(train_txt_path, 'w')
for i in range(number_images_trainset):
    train_txt.write(list_images[i] + '\n')
train_txt.close()

val_txt_path = "/workspace/nabang1010/STEAM/PhamMinhDuc_STEAM/src/yolov7/valset.txt"
val_txt = open(val_txt_path, 'w')
for i in range(number_images_trainset, number_images):
    val_txt.write(list_images[i] + '\n')
val_txt.close()
```
